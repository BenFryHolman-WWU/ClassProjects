```python
"""
Please forgive this implementaion
"""
def unhyphen(input):
  if input == np.nan:
    return np.nan
  elif 'µ' in str(input):
    return 0
  elif 'k' in str(input):
    return float(str(input)[:-1])*1000
  elif 'M' in str(input):
    return float(str(input)[:-1])*1000000
  elif 'B' in str(input):
    return float(str(input)[:-1])*1000000000
  elif input == 'NaN':
    return np.nan
  elif isinstance(input, str):
    output = ''
    for x in input:
      if x == '−':
        output += '-'
      else:
        output += x
    return float(output)
  else:
    return input
```

```python
for x in gdp.columns:
  for y in gdp[str(x)]:
    gdp[str(x)] = gdp[str(x)].apply(lambda x : unhyphen(x))
```

```python
def imputeByRowNumber(number):
  imp = SimpleImputer(missing_values=np.nan, strategy='mean')
  x = imp.fit_transform(gdp.iloc[number].values.reshape(1, -1))[0]
  difference = gdp.shape[1] - len(x)
  for y in range(difference):
    x = np.insert(x, 0, 0)
  gdp.iloc[number] = x

```

```python
for x in range(len(gdp)):
  imputeByRowNumber(x)
```
